```python
# 1 .import the necessary libraries
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
# 2. Import the dataset from the address given
df = pd.read_csv('wine.data')
```

```python
df.head()
```

```python
# 3. Assign it to a varaible called wine
wine = df
```

```python
# 4. Delete the first, fourth, seventh, nineth, eleventh, thirteenth and fourteenth columns

wine.drop(wine.columns[[0,3,6,8,10,12,13]],axis=1)
```

```python
# 5. Assign the column names
wine.columns=['Type','Alcohol','Malic acid','Ash','Alcalinity of ash','Magnesium','Total phenols','Flavanoids','Nonflavanoid phenols','Proanthocyanins','Color intensity','Hue','OD280/OD315 of diluted wines','Proline']
```

```python
wine.head()
```

```python
# 6. Set the values of the first 3 rows from alcohol as NaN
wine['Alcohol'][0:3] = np.nan
wine.head()
```

```python
# 7. Now set the value of the rows 3 and 4 of magnesium as NaN
wine['Magnesium'][2:4] = np.nan
wine.head()
```

```python
# 8. Fill the value of NaN with the number 10 in alcohol and 100 in magnesium
wine['Alcohol'].replace(np.nan, 10, inplace= True)
wine['Magnesium'].replace(np.nan,100, inplace = True)
wine.head()
```

```python
# 9. Count the number of missing values
wine.isnull().sum()
```

```python
# 10. Create an array of 10 random numbers up until 10
import random
res =[]
for i in range(10):
    res.append(random.randint(0,10))
res
```

```python
# 11.Use random numbers you generated as an index and assign NaN value to each of cell.
wine.loc[res]=np.nan
wine.head(10)
```

```python
# 12. How many missing values do we have?
wine.isnull().sum()
```

```python
# 13. Delete the rows that contain missing values
wine.dropna(inplace=True)
```

```python
# 14. Print only the non-null values in alcohol
wine['Alcohol'][wine['Alcohol'] != np.nan]
```

```python
# 15. Reset the index, so it starts with 0 again
wine.reset_index()
```
